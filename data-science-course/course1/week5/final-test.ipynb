```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import sqrt
```

```python
b = 3

pareto_rv = sts.pareto(b)

mean = sts.pareto.mean(b)
variance = sts.pareto.var(b)


# generates 'number_of_samples' samples of size 'samples_size' and returns an array of their means
def f(sample_size, number_of_samples):
    result = np.asarray([])
    for i in range(number_of_samples):
        result = np.append(result, np.mean(pareto_rv.rvs(sample_size)))
    return result
```

```python
sample = pareto_rv.rvs(1000)
plt.hist(sample, bins=100, normed=True, label='sample')

x = np.linspace(1, 10, 100)
pdf = pareto_rv.pdf(x)
plt.plot(x, pdf, color='r', label='theoretical PDF')

plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
```

# N = 5

```python
n = 5

plt.hist(f(n, 1000), bins=20, normed=True, label='sample')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

norm_rv = sts.norm(mean, sqrt(variance / n))
x = np.linspace(0, 6, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf, color='r', label='theoretical PDF')
plt.legend()
```
