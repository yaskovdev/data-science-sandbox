```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

```python
x = np.arange(1.0, 15.0, 0.01)


def f(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)
```

```python
a1 = np.array([[1.0,1.0],[1.0,15.0]])
b1 = np.array([f(1.0), f(15.0)])
```

```python
linalg.solve(a1, b1)
```

```python
def polynomial1(x):
    return 3.43914511 - 0.18692825 * x
```

```python
plt.plot(x, f(x), x, polynomial1(x))
```

```python
a2 = np.array([[1, 1, 1], [1, 8, 16], [1, 15, 225]])
b2 = np.array([f(1.0), f(8.0), f(15.0)])
```

```python
linalg.solve(a2, b2)
```

```python
def polynomial2(x):
    return 3.38154959 - 0.12549303 * x - 0.0038397 * x ** 2
```

```python
plt.plot(x, f(x), x, polynomial2(x))
```
