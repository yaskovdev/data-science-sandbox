```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, m = 3., 2.
s1000 = (np.random.pareto(a, 1000) + 1) * m

count, bins, _ = plt.hist(s1000, 100, normed=True)
fit = a * m ** a / bins ** (a + 1)
plt.plot(bins, max(count) * fit / max(fit), linewidth=2, color='r')
plt.show()
```

# N = 5

```python
def f5():
    sum = (np.random.pareto(a, 5) + 1) * m
    print(sum)
    for i in range(100):
        sum += (np.random.pareto(a, 5) + 1) * m
    return sum / 101


count, bins, _ = plt.hist(f5(), 100, normed=True)
fit = a * m ** a / bins ** (a + 1)
plt.plot(bins, max(count) * fit / max(fit), linewidth=2, color='r')
```
