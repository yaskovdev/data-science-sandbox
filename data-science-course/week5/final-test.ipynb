```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
a = 3.
s1000 = (np.random.pareto(a, 1000) + 1)

count, bins, _ = plt.hist(s1000, 100, normed=True)
fit = a * 1 ** a / bins ** (a + 1)
plt.plot(bins, max(count) * fit / max(fit), linewidth=2, color='r')
plt.show()
```

```python
b = 3

pareto_rv = sts.pareto(b)

mean = sts.pareto.mean(b)
variance = sts.pareto.var(b)


def f(samples_size, number_of_samples):
    result = np.zeros(samples_size)
    for i in range(number_of_samples):
        result += pareto_rv.rvs(samples_size)
    return result / number_of_samples
```

# N = 50

```python
n = 20

plt.hist(f(n, 1000), bins=20, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

norm_rv = sts.norm(mean, variance / n)
x = np.linspace(1.425, 1.625, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf, color='r')
```
