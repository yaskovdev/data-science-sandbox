```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
b = 3

pareto_rv = sts.pareto(b)

mean = sts.pareto.mean(b)
variance = sts.pareto.var(b)


# generates 'number_of_samples' samples of size 'samples_size' and returns 
def f(sample_size, number_of_samples):
    result = np.zeros(sample_size)
    for i in range(number_of_samples):
        result += pareto_rv.rvs(sample_size)
    return result / number_of_samples
```

```python
sample = pareto_rv.rvs(1000)
plt.hist(sample, bins=100, normed=True)

x = np.linspace(1, 10, 100)
pdf = pareto_rv.pdf(x)
plt.plot(x, pdf, color='r')

plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

# N = 50

```python
n = 20

plt.hist(f(n, 1000), bins=20, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

norm_rv = sts.norm(mean, variance / n)
x = np.linspace(1.425, 1.625, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf, color='r')
```
