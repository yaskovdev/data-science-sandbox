```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, m = 3., 2.
s1000 = (np.random.pareto(a, 1000) + 1) * m

count, bins, _ = plt.hist(s1000, 100, normed=True)
fit = a * m ** a / bins ** (a + 1)
plt.plot(bins, max(count) * fit / max(fit), linewidth=2, color='r')
plt.show()
```

```python
import scipy.stats as sts
```

```python
pareto_rv = sts.pareto(3)


def f(n, size):
    result = np.zeros(n)
    for i in range(size):
        result += pareto_rv.rvs(n)
    return result / size
```

```python
plt.hist(f(50, 1000), bins=40, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
plt.hist(f(100, 1000), bins=30, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```
